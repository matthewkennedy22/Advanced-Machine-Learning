## Part 2 — Cost of living and salary

I picked 3 cities and guessed monthly costs. Change the numbers in the code if you want.


```python
import pandas as pd
import matplotlib.pyplot as plt

# inputs you can change
tax_keep = 0.72          # rough: how much of your salary you keep after taxes
want_extra = 0.20        # want target salary this much above minimum
city_a = "Austin, TX"
city_b = "Denver, CO"
city_c = "Atlanta, GA"

df = pd.DataFrame({
    "city": [city_a, city_b, city_c],
    "housing": [1850, 2100, 1650],
    "food": [550, 600, 520],
    "transport": [350, 320, 400],
    "utilities": [180, 200, 170],
    "other": [400, 450, 380],
})

df["monthly_total"] = (
    df["housing"] + df["food"] + df["transport"] + df["utilities"] + df["other"]
)

# min salary = need enough gross so after tax you cover 12 months of spending
df["min_salary"] = df["monthly_total"] * 12 / tax_keep

# desired salary = a bit higher for savings / buffer
df["target_salary"] = df["min_salary"] * (1 + want_extra)

print(df)

plt.figure(figsize=(7, 4))
plt.bar(df["city"], df["monthly_total"])
plt.ylabel("dollars per month")
plt.title("total monthly cost by city")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```
